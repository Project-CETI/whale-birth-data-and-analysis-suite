## Load all segmentation files

```python
import glob 
```

```python
clean_seg_files = glob.glob("../data/*cleaned.pkl")
raw_seg_files = glob.glob("../data/*/1688*.hdf5")

clean_seg_files = sorted(clean_seg_files)
```

```python
raw_seg_files
```

```python
vid_numbers_raw = [file_name.split("/")[-1].split("_")[0] for file_name in raw_seg_files]
sorted_vid_idcs = np.argsort(vid_numbers_raw)
raw_seg_files = list(np.array(raw_seg_files)[sorted_vid_idcs])
#vid_numbers = list(np.sort(vid_numbers))

# the following are already sorted
vid_numbers = [file_name.split("/")[-1][:-12] for file_name in clean_seg_files]

vid_numbers
```

```python
vid_number2file = {i : j for i in vid_numbers for j in raw_seg_files if i[:13] in j}
vid_number2file
```

```python
import pickle 

segmentations = {}

# load clean segmentations
for vid_number in vid_numbers:

    file_name = f"../data/{vid_number}_cleaned.pkl"
    with open(file_name, "rb") as fp:
        segmentation = pickle.load(fp)

        segmentations[vid_number] = segmentation
```

## Load masks

```python
# load masks from raw segmentations
for i, vid_number in enumerate(vid_numbers):

    print(vid_number)
    seg = segmentations[vid_number]
    
    keep_indices = seg["frame_indices"]
    seg_original = Segmentations(h5_filepath=vid_number2file[vid_number], video_filepaths=None)
    print("loading masks")
    masks = seg_original.get_all_masks_contours()
    masks_filtered = masks[keep_indices]

    print("adding polygons")
    seg["masks_arrays"] = masks_filtered
    seg = add_masks_polygons(seg)

    segmentations[vid_number] = seg
```

```python
import pickle 

for vid_number in vid_numbers:

    file_name = f"../data/{vid_number}_cleaned_with_masks.pkl"
    with open(file_name, "wb") as fp:
        pickle.dump(segmentations[vid_number], fp)
```

### Visualize

```python

for vid_number in vid_numbers:

    seg = segmentations[vid_number]


    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 2))
    plot_presence(seg, ax=ax1)
    ax2.plot(seg["frame_indices"], seg["num_whales_over_time"], 'o')
    ax1.set_title(f"vid {vid_number}")

plt.show()
```
