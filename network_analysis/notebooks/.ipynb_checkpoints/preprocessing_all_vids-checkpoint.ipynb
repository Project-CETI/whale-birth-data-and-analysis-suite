## Load all segmentation files

```python
data_files = glob.glob("../data/*/1688*.hdf5")
data_files
```

```python
import os 
drone_data_dir = "../data/drones/"

baby_epoch_s = time_str_to_time_s('2023-07-08 11:45:45 -0400')

drone_data_hdf5_filepaths = {
  'CETI': os.path.join(drone_data_dir, 'CETI-DJI_MAVIC3-1_metadata.hdf5'),
  'DSWP': os.path.join(drone_data_dir, 'DSWP-DJI_MAVIC3-2_metadata.hdf5'),
}

# load drone data containing frame times
drone_videos = DroneVideos(
    drone_data_hdf5_filepaths=drone_data_hdf5_filepaths,
    video_dirs=None,
    custom_epoch_time_s=baby_epoch_s
)
```

```python
segs_raw = {}

for file_name in data_files:
    
    vid_number = file_name.split("/")[-1].split("_")[0]
    print(vid_number)
    # load the segmentations data
    segmentations = Segmentations(h5_filepath=file_name, video_filepaths=None)

    # load frame times from drone data
    timestamps_s = drone_videos.get_frame_timestamps_s(video_key=vid_number, use_custom_epoch=False) # standard posix epoch time
    timestamps_s_babyTime = drone_videos.get_frame_timestamps_s(video_key=vid_number, use_custom_epoch=True) # baby time!
    timestamps_str = drone_videos.get_frame_timestamps_str(video_key=vid_number)

    frame_times = [timestamps_s, timestamps_s_babyTime, timestamps_str]

    #extract the dict from hdf5 file
    seg_raw = convert_to_dict(segmentations, frame_times=frame_times, with_masks=False)

    if vid_number=="1688841618482": # duplicate 1688841618482 because has two separate parts
        segs_raw[f"{vid_number}_1"] = seg_raw
        segs_raw[f"{vid_number}_2"] = seg_raw
    else : 
        segs_raw[vid_number] = seg_raw
```

```python
vid_numbers = list(segs_raw.keys())
vid_numbers
```

### Visualize

```python
for vid_number in vid_numbers:

    seg_raw = segs_raw[vid_number]
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 2))
    plot_presence(seg_raw, ax=ax1)
    ax2.plot(seg_raw["num_whales_over_time"], 'o')
    ax1.set_title(f"vid {vid_number}")
```

```python
frame_lims = {}

#for number in vid_numbers:
#    frame_lims[number] = [None, None] # min and max frames to consider 

# those limits are chosen based on the videos and the above visualizations
frame_lims["1688841618482_1"] = [None, 2500]
frame_lims["1688841618482_2"] = [3800, 5000]

frame_lims["1688827660979"] = [None, None]
frame_lims["1688830733272"] = [0, 5000]
frame_lims["1688829151574"] = [None, None]
frame_lims["1688830960531"] = [1000, 5000]
frame_lims["1688827433752"] = [2600, 6400]
frame_lims["1688828965144"] = [1250, 6000]
frame_lims["1688827887072"] = [0, 1000]
frame_lims["1688829193016"] = [0, 3000]
frame_lims["1688831602631"] = [None, None]
frame_lims["1688832767759"] = [None, None]
frame_lims["1688831830406"] = [None, None]
frame_lims["1688832995052"] = [None, None]
frame_lims["1688832540499"] = [None, None]
frame_lims["1688831374924"] = [None, None]
frame_lims["1688831228731"] = [None, None]
frame_lims["1688831187858"] = [None, None]

import json

# we can use this file for batch preprocessing videos in consistent way
file_name = "../data/frame_limits.json"
with open(file_name, 'w') as fp:
    json.dump(frame_lims, fp, indent=4)
```

### Clean: cut after/before frame min/max, remove non-segmented frames, smooth positions/orientations

```python
from utils_cleaning import slice_seg, remove_nonsegmented_frames
```

```python
segs = {}

window = 100 # for median smoothin in frames

for vid_number in vid_numbers:
    print(vid_number)

    vmin, vmax = frame_lims[vid_number]

    seg = segs_raw[vid_number]

    # ajust rolling window based on segmentation_frequency
    if seg["segmentation_frequency"][0] == 20:
        window = 10
    elif len(seg["segmentation_frequency"]) > 1:
        window = 5

    # start cleaning
    seg = slice_seg(seg, frame_min=vmin, frame_max=vmax)
    seg = remove_nonsegmented_frames(seg)
    seg = smooth_seg_dict(seg, window=window)
    
    segs[vid_number] = seg
```

```python
for vid_number in vid_numbers:

    seg = segs[vid_number]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 2))
    plot_presence(seg, ax=ax1)
    ax2.plot(seg["frame_indices"], seg["num_whales_over_time"], 'o')
    ax1.set_title(f"vid {vid_number}")
```

```python
import pickle 

for vid_number in vid_numbers:

    file_name = f"../data/{vid_number}_cleaned.pkl"
    with open(file_name, "wb") as fp:
        pickle.dump(segs[vid_number], fp)
```
